```python
# pip3 install numpy
# pip3 install pandas
# pip3 install matplot
# https://matplotlib.org/stable/index.html 

import numpy as np
import pandas as pd 
import matplotlib 

#50개 랜덤한 수 생성 
data = np.random.rand(50)
print(type(data))
print(f"출력해보기 data : {data} ")

# 넘파이 
# 넘파이 -> 판다스의 시리즈 자료형으로 변환 
seri = pd.Series(data)
print(type(seri))
print(f"출력해보기 seri : \n {seri} ")

# 판다스는 == 테이블
# 변환된 pd(테이블)
# 변환된 pd(테이블), 차트 그리기. 
#matplotlib 으로 스타일변경해보개 
matplotlib.style.use("ggplot")
seri.plot()

```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot:
fig, ax = plt.subplots()

ax.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))

ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.show()
```
