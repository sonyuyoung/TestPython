```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot:
fig, ax = plt.subplots()

ax.hist2d(x, y, bins=(np.arange(-3, 3, 0.1), np.arange(-3, 3, 0.1)))

ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.show()
```

```python
#np 슬라이싱 
import numpy as np
import pandas as pd 
import matplotlib 
# 3x2 행렬
# 1 2
# 3 4
# 5 6
np_array= np.array(list)
list =[[1,2],[3,4],[5,6]]
# 2번째 행(인덱스 1)의 2번째 열(인덱스 1)의 요소를 선택
print(f"np_array[1][1] : {np_array[1][1]}")
#  1번째 행(인덱스 0)부터 2번째 행(인덱스 1) 전까지를 선택 
#  주의할 점은 범위의 끝 인덱스는 포함되지 않는다
print(f"np_array[0:1] : {np_array[0:1]}")
# 리스트에서 1번째 행(인덱스 0)부터 3번째 행(인덱스 2) 전까지를 선택
print(f"np_array[0:2] : {np_array[0:2]}")
# 전체행 1열 
print(f"np_array[:,1] : {np_array[:,1]}")
# np_array[[0,2],1] 0행 2행의 1열 부분 ,2,6
print(f"np_array[[0,2],1] : {np_array[[0,2],1]}")
```

```python
# 브로드 캐스트 정렬 
import numpy as np
import pandas as pd 
import matplotlib 
# 3x2 행렬
# 1 2
# 3 4
# 5 6
np_array= np.array(list)
# 각 넘파이 배열에 모든 요소에 특정값 한번에 모두 더하기 : 브로드캐스트 : 전체적으로 다 전달? 
print(f"np_array 연산전 : \n{np_array}")
print(f"np_array +100 연산후 : \n{np_array+100}")

# 각행 열의합 axis=0 열기준 , 1은 행기준 
# 넘파이는 1행 이다. 1행기준 
print(f"np_array 0열기준의 합 : \n{np_array.sum(axis=0)}")
print(f"np_array 0행기준의 합 : \n{np_array.sum(axis=1)}")

# 각 행, 열의 기준 정렬 
print(f"np_array 행 기준 정렬: \n {np.sort(np_array, axis=1)} ")
print(f"np_array 열 기준 정렬: \n {np.sort(np_array, axis=0)} ")
print(f"np_array 열 기준  내림차순 정렬: \n {np.sort(np_array, axis=0)[::-1]} ")
```

```python
# 판다스의 시리즈에서 , 리스트 컬럼스 옵션 zip 함수이용 
# 07 -> 26p
import numpy as np
import pandas as pd 
import matplotlib

data= [["유영","유영2","유영3"],
       ["메뉴1","메뉴2","메뉴3"],
       ["1000","2000","3000"]]

#  옵션 zip 함수이용 같은 인덱스 묶기
data2 = zip(data[0],data[1],data[2])
df=pd.DataFrame(data2,columns=["이름","메뉴","가격"])
print(f"df:\n {df}")

# 각열 선택 
print(f"df이름열선택:\n {df["이름"]}")

# 각 열값 대입
df["이름"] = '유영'
print(f"df이름열값 수정:\n {df["이름"]}")
#삭제 
del df["이름"]
print(f"df  이름 열 값 삭제 후 조회 : \n {df} ")

# 행선택 수정삭제 
print(f"df.loc[:] 전체 행 조회 : \n {df.loc[:]} ")
print(f"df.loc[0:1] 전체 행 조회 : \n {df.loc[0:1]} ")

df.loc[0]="10000000000"
print(f"df.loc[0]0행 값수정  : \n {df} ")

# 특정행에 값추가 
# 행의 이름으로도 접근이 가넝  
df.loc[0,:]=["test1","test2"]
print(f"df.loc[0,:] 0행값 추가  : \n {df} ")

# iloc , 해당 인덱스읭 값으로 선택, 변경이 가능 
```
